```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.iloc[tips["tip"].argmax()]
```

```python
sns.barplot(tips,x='sex',y='tip',estimator='max')
plt.show()
```

```python
sns.barplot(tips,x='time',y='tip',hue='day',estimator='mean')
plt.show()
```

```python
sns.histplot(tips,x="tip")
```

```python
sns.histplot(tips,x="tip",hue="sex")
```

```python
sns.kdeplot(tips,x="tip")
```

```python
sns.kdeplot(tips,x="tip",hue="sex")
```

```python
sns.scatterplot(tips,x="total_bill",y="tip",hue="day")
```

```python
sns.scatterplot(tips,x="total_bill",y="tip",hue="time")
```

```python
sns.scatterplot(tips,x="total_bill",y="tip",hue="sex")
```
